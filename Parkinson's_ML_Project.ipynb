```python
# Import libraries
import numpy as np
import pandas as pd
import os, sys
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Read in the dataset
df = pd.read_csv(r"C:\Users\adeel\Downloads\parkinsons.txt")
df.to_csv(r"C:\Users\adeel\Downloads\parkinsons.txt", index = False)
df.head()
```

```python
# Get features and labels
features = df.loc[:, df.columns != 'status'].values[:, 1:]
labels = df.loc[:, 'status'].values
```

```python
# Get counts of each label (0 and 1) in labels
print(labels[labels == 0].shape[0], labels[labels == 1].shape[0]) 
```

```python
# Initialize a MinMaxScaler and scale features between -1 and 1
scaler = MinMaxScaler((-1, 1))
x = scaler.fit_transform(features)
y = labels
```

```python
# Split dataset into training and testing, keeping 20% of the data for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 7)
```

```python
# Initialize XGBClassifier and train the model
model = XGBClassifier()
model.fit(x_train, y_train)
```

```python
# Generate predicited values for x_test and calculate model accuracy
y_pred = model.predict(x_test)
print(accuracy_score(y_test, y_pred) * 100)
```
